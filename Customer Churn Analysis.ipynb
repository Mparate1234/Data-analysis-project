```python
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Step 2: Get percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 3: Plot stacked bar chart
ax = percentages.plot(kind='bar', stacked=True, figsize=(8, 5))

# Step 4: Add percentage labels
for container in ax.containers:
    labels = [f'{v.get_height():.1f}%' if v.get_height() > 0 else '' for v in container]
    ax.bar_label(container, labels=labels, label_type='center', color='Black', fontsize=10)

# Step 5: Customize plot
plt.title("Customer Churn % by Senior Citizen", color="w")
plt.xlabel("Senior Citizen",color="w")
plt.ylabel("Percentage",color="w")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Layout settings
n_cols = 3
n_rows = -(-len(cols) // n_cols)  # Ceiling division

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 12), facecolor='black')
axes = axes.flatten()

# Loop through each column and plot
for i, col in enumerate(cols):
    ax = axes[i]
    sc = sns.countplot(data=df, x=col, ax=ax, palette='viridis', hue =df["Churn"])

    # Set white text and style
    ax.set_title(f'Countplot of {col}', color='white')
    ax.tick_params(axis='x', rotation=45, colors='white')
    ax.tick_params(axis='y', colors='white')
    ax.set_xlabel('', color='white')
    ax.set_ylabel('Count', color='white')

    # Set spines color to white
    for spine in ax.spines.values():
        spine.set_color('white')

    # Add count labels in white
    for container in sc.containers:
        sc.bar_label(container, fmt='%d', color='white', fontsize=10)

# Remove any extra axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Set super title
plt.suptitle("Service Feature Countplots", y=1.02, fontsize=16, color='white')

# Optional: Set dark background
fig.patch.set_facecolor('black')
plt.tight_layout()
plt.show()
```
